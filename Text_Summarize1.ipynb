```python
from youtube_transcript_api import YouTubeTranscriptApi

srt = YouTubeTranscriptApi.get_transcript("UHRPtNZ_Rz4")
print (srt)
print('\n')
para = ""

for i in srt:
	para += i['text'].replace('\n',"") + " "

import requests

url = 'http://bark.phon.ioc.ee/punctuator'


unpunctuated_text = {
  'text': para
}

response = requests.post(url, data = unpunctuated_text)
print(para)
print('\n')
print(response.text)
text_=response.text
```

**Generate Summary**

```python
!pip install -U spacy
!python -m spacy download en_core_web_sm
```

Import libraries

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

### For Text Cleaning

```python
stopwords=list(STOP_WORDS)
print(stopwords)
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(text_)
```

```python
print(doc)
```

### Word Tokenization

```python
tokens=[token.text for token in doc]
print(tokens)
```

```python
 
punctuation
```

Code to generate word frequencies where stopwords and punctuations are not considered

```python
word_frequencies={}
for word in doc:
  if word.text.lower() not in stopwords:
      if word.text.lower() not in punctuation:
          if word.text not in word_frequencies.keys():
              word_frequencies[word.text]= 1
          else:
              word_frequencies[word.text]+=1
```

### Word Frequencies

```python
print(word_frequencies)
```

### Normalized frequency

```python
max_frequency= max(word_frequencies.values())
```

```python
max_frequency
```

```python
for word in word_frequencies.keys():
  word_frequencies[word]=word_frequencies[word]/max_frequency
```

```python
print(word_frequencies)
```

Sentence Tokenization

```python
sentence_tokens=[sent for sent in doc.sents]
print(len(sentence_tokens))
print(list(sentence_tokens))
```

Code to generate score for each sentence based on frequencies of words in the sentence.

```python
sentence_scores={}
for sent in sentence_tokens:
  for word in sent:
    if word.text.lower() in word_frequencies.keys():
      if sent not in sentence_scores.keys():
        sentence_scores[sent]=word_frequencies[word.text.lower()]
      else:
        sentence_scores[sent]+=word_frequencies[word.text.lower()]
```

```python
sentence_scores
```

```python
from heapq import nlargest
```

To get the 20% of the complete summary text

```python
select_length =int(len(sentence_tokens)*0.20)
select_length
```

To get sentences with maximum score

```python
summary=nlargest(select_length,sentence_scores,key=sentence_scores.get)
```

```python
summary
```

```python
final_summary=[word.text for word in summary]
final_summary
```

Convert array of sentences to a paragraph of sentences

```python
summary=' '.join(final_summary)
```

```python
summary
```

```python
summary=summary.replace('\n','')
```
